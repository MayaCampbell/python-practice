We have 36 starship records
10 records per page
total of pages 4

We need to get all of the pages into one Python List.
Pseudo-code:

```python
baseurl= "https://swapi.dev/api/starships/?page="
response= requests.get(baseurl)
view_response= response.json()

#print(view_response)

totalpg= 4
#Getting all pages into 1 list
data=[] #Step 1: create empty list
for i in range(1, totalpg+1): #Step 2: for loop iterating over range of pages 
    response= requests.get(baseurl + str(i)) #Step 3: get request concatenating url and page# as string and assigning to a variable
    new_response= response.json()#Step 4: .json() to view json file and assign to new variable
    data.append(new_response)#Step 5: append new response to data list

print(data)#YAY! Now all the pages are in one list


```

Great! Now we want to iterate of this list only through the results key.

```python
cleaned_data= [] #Step 1: create empty list

for list_pages in data: #Step 2: iterate over each 'page' in data list
    for values in list_pages['results']: #Step 3: iterate over each 'value in 'page at 'results' index
        cleaned_data.append(values) #Setp 4: append values to cleaned data list

print(cleaned_data)
```

3. Create a function that returns a list of all the starships that appear in more than x number of films. Where 'x' is the number of films you are checking for and is passed into the function.
Pseudo-Code:
1. create empty list
2. loop through name key in cleaned data

```python
def num_films(x):
    film_num= []
    for name in cleaned_data:
        if len(name['films']) > x:
            film_num.append(name['name'])
    return film_num

num_films(1)
```

4. Create function that returns a tuple of the min, max, and avg length for all starship names

```python
def find_length():
    lengths= []
    lengths_cleaned= []
    for length in cleaned_data:
        lengths.append(length['length'])
        for length in lengths:
            if length.isnumeric():
                lengths_cleaned.append(int(length))
    lengths_cleaned.sort()
    return min(lengths_cleaned), max(lengths_cleaned), sum(lengths_cleaned)//len(lengths_cleaned)
    
find_length()

```
